在本次项目中, 我们小组围绕"知识管理系统"这一课题, 做了C/S架构应用开发以及NLP/CV研发的两方面工作. 机会难得, 收获颇丰, 总结如下:
* 搭建了基于Client/Server架构的内容类应用, 在开发过程中, 小组成员都积累了一定的开发经验和知识储备.
* 围绕"知识管理"的目的, 主要在NLP和CV两方向做了很多调研和实践, 为我们的产品赋能. 在调研过程中, 我们接触了很多领域的知识, 积累了一些科研经验.
* 整个项目过程, 队员们分工合理明确, 都对产品做出了一定的贡献. 不论是一起研讨研发上的技术选型, 还是开发过程中的分工到每个人, 都帮助我们积累了大量团队合作经验.

经过反思, 本项目也存在一些不足和尚待进一步研究的地方, 整理如下:
* 选题宽泛, 在技术选型, 尝试, 到最终落地的过程耗费了过长时间. 在选题初期, 我们花了很多时间进行课题调研, 虽然是一段稳定的学习时期, 但由于没有一个产品雏形作为指导, 积极性不如中后期为作品添砖加瓦的阶段, 也走了不少弯路.
* 调研中发现很多领域知识需要较长时间的学习, 结合第一点, 我们最终决定技术上重沉淀而不贪多, 产品功能上只选用经过反复学习调研, 了解细节的功能.
* 在NLP领域的调研学习中, 我们发现中文语料库的匮乏会遏制很多有趣想法的开展, 调研初期加入蓝图的一些功能最终被弃用或是用传统方法实现.

总的来说, 虽然在选题到落地到产品成型的过程中经历过迷茫的时期, 我们最终还是呈现出一个功能完整闭环的产品, 且在一次次学中做, 做中学的过程中加深了对选题的理解和对相应领域知识的掌握.
